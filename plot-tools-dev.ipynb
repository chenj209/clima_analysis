```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def create_plot_object():
    fig, ax = plt.subplots()
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    ax.plot(x, y)
    return fig, ax

def integrate_subplots():
    # 创建一个大的画布，包含2x2的子图
    figs, axs = plt.subplots(2, 2, figsize=(10, 10))
    
    # 使用 create_plot_object 函数创建四个子图
    for i in range(2):
        for j in range(2):
            _, ax_temp = create_plot_object()
            axs[i, j].plot(ax_temp.lines[0].get_xdata(), ax_temp.lines[0].get_ydata())
    
    plt.tight_layout()
#     plt.show()
    
    return figs, axs
```

```python
FIG, AXS = integrate_subplots()
FIG.savefig("my_figure.png")
```
